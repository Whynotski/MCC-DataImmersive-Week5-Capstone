```python
knn_name = ['Diff-Walk across room', 
     'Diff-Walk sev blocks',
       'Diff-Walk one block', 'Diff-Sit for 2 hours', 'Diff-Get up fr chair',
       'Diff-Climb sev flt stair', 'Diff-Climb one flt stair', 'bend',
       'Some diff walking', 'Some diff walking one block',
       'Difficulty several floors', 'Some Diff-Clmb 1 flt stair',
       'Some Diff-Walk across room']    
            
      
```

```python
#make a dataframe called X to hold the predictive features
mob_col = pd.DataFrame(df,columns=knn_name)
mob_col.head()
```

```python
mob_col.dtypes
```

```python
miss_dict = {"0.no":0, "1.yes": 1, ".d":np.NAN, ".r":np.NAN, ".s":np.NAN, ".x":np.NAN, "2.can't do":0, "9.don't do":1}
```

```python
mob_col2 = mob_col.replace(miss_dict)
```

```python
mob_col2.dtypes
```

```python
mob_col2.describe()
```

```python
X = mob_col2.dropna()
```

```python
X.describe()
```

```python
#initialize k-means function
#set number of clusters to categorize = 4
kmeans = KMeans(n_clusters=3)

#the model will learn which data points seem similar
#calculates centroids
#classfies data points in clusters based on distance to centroids
kmeans.fit(X)
```

```python
np.set_printoptions(suppress=True,precision=1)
```

```python
#data point coordinates for centroids of each cluster
#index 0 is blue cluster
#index 1 is yellow cluster
#index 2 is purple cluster
#index 3 is red cluster
print(kmeans.cluster_centers_)
```

```python
#add a new column to dataframe called "cluster"
#can use this to compare features of clusters to original categories
X['cluster'] = kmeans.labels_
```
